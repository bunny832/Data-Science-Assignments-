```python
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
for col in ['LB', 'AC', 'FM', 'UC', 'DL', 'DS', 'DP', 'ASTV', 'MSTV', 'ALTV', 'MLTV', 'Width', 'Tendency', 'NSP']:
    plt.figure(figsize=(8,4))
    sns.boxplot(x=col,data=df)
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

summary = df.mean()

# Increase figure size
plt.figure(figsize=(12, 8))

# Create the bar plot with rotated x-axis labels
summary.plot(kind='bar')

# Rotate x-axis labels for better readability
plt.xticks(rotation=55, ha='right')

# Add title and labels
plt.title('Mean Values of Different Features')
plt.xlabel('Features')
plt.ylabel('Mean Value')

# Show the plot
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
# If you want to plot histograms for all numerical columns, you can do:
columns = df.select_dtypes(include=['float64', 'int64']).columns

# Set up the figure
plt.figure(figsize=(15, 10))

# Loop through the selected columns and create a histogram for each
for i, column in enumerate(columns):
    plt.subplot(len(columns) // 3 + 1, 3, i + 1)  # Create subplots
    df[column].hist(bins=20)  # Plot histogram with 20 bins
    plt.title(column)  # Add the title as the column name

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plots
plt.show()
```

```python
# Assuming you want to create scatter plots for the first few numerical columns
columns = df.select_dtypes(include=['float64', 'int64']).columns

# Set up the figure
plt.figure(figsize=(15, 10))

# Loop through the selected columns in pairs and create a scatter plot for each
for i in range(0, len(columns) - 1, 2):  # Step by 2 to get pairs
    plt.subplot(len(columns) // 2, 2, i // 2 + 1)  # Create subplots
    plt.scatter(df[columns[i]], df[columns[i + 1]])
    plt.title(f'{columns[i]} vs {columns[i + 1]}')
    plt.xlabel(columns[i])
    plt.ylabel(columns[i + 1])

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plots
plt.show()
```
