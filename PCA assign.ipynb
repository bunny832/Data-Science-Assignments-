```python
# prompt: how to remove outliers to the above dataset in a short code

for column in df.columns:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Standardize the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df)

# Apply PCA
pca = PCA(n_components=14) # You can choose the number of components
principal_components = pca.fit_transform(scaled_features)

# Create a new DataFrame with the principal components
pca_df = pd.DataFrame(data = principal_components, columns = ['principal_component_1', 'principal_component_2','principal_component_3','principal_component_4','principal_component_5','principal_component_6','principal_component_7','principal_component_8','principal_component_9','principal_component_10','principal_component_11','principal_component_12','principal_component_13','principal_component_14'])

print(pca_df.head())
```

```python
import matplotlib.pyplot as plt
# Determine the optimal number of components using a scree plot
plt.figure(figsize=(10, 6))
plt.plot(range(1, pca.n_components_ + 1), pca.explained_variance_ratio_, marker='o', linestyle='--')
plt.title('Scree Plot')
plt.xlabel('Number of Principal Components')
plt.ylabel('Explained Variance Ratio')
plt.xticks(range(1, pca.n_components_ + 1))
plt.grid(True)
plt.show()

# Determine the optimal number of components using cumulative explained variance
cumulative_explained_variance = pca.explained_variance_ratio_.cumsum()
plt.figure(figsize=(10, 6))
plt.plot(range(1, pca.n_components_ + 1), cumulative_explained_variance, marker='o', linestyle='--')
plt.title('Cumulative Explained Variance')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance Ratio')
plt.xticks(range(1, pca.n_components_ + 1))
plt.axhline(y=0.95, color='r', linestyle='-', label='95% Explained Variance') # Example threshold
plt.legend()
plt.grid(True)
plt.show()

# Based on the plots, you can choose the number of components that capture a sufficient amount of variance
# For example, finding the number of components for 95% explained variance
optimal_components = next(i for i, cumulative_variance in enumerate(cumulative_explained_variance) if cumulative_variance >= 0.95) + 1
print(f"Number of components to explain at least 95% variance: {optimal_components}")
```

```python
# For now, principal_components holds the transformed data with 14 components.
# If you want to assign this to a new variable for clarity or further use:
transformed_data = principal_components

# You can print the shape of the transformed data to verify
print("Shape of the transformed data:", transformed_data.shape)

# You can also print the first few rows of the transformed data (same as pca_df head)
print("\nFirst 5 rows of the transformed data:")
print(pd.DataFrame(transformed_data, columns = [f'principal_component_{i+1}' for i in range(transformed_data.shape[1])]).head())
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Determine the optimal number of clusters using the elbow method
inertia = []
range_n_clusters = range(1, 11) # You can adjust the range of clusters to test
for num_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=num_clusters, random_state=42, n_init=10) # Explicitly set n_init
    kmeans.fit(df)
    inertia.append(kmeans.inertia_)
    
# Let's assume we choose k=3 as an example (adjust based on your plot)
optimal_clusters = 3

# Apply K-means clustering to the original dataset with the chosen number of clusters
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42, n_init=10) # Explicitly set n_init
clusters = kmeans.fit_predict(df)

# Add the cluster labels to the original DataFrame
df['cluster'] = clusters

# You can now analyze the clusters, e.g., by looking at the mean of features for each cluster
print("\nMean of features per cluster:")
print(df.groupby('cluster').mean())

# You can also visualize the clusters if you have a manageable number of dimensions,
# or use PCA to visualize clusters in lower dimensions.
# For example, if you used PCA previously and have 'pca_df':
if 'principal_component_1' in pca_df.columns and 'principal_component_2' in pca_df.columns:
    plt.figure(figsize=(10, 8))
    sns.scatterplot(x='principal_component_1', y='principal_component_2', hue=clusters, data=pca_df, palette='viridis', legend='full')
    plt.title('Clusters visualized on first two Principal Components')
    plt.xlabel('Principal Component 1')
    plt.ylabel('Principal Component 2')
    plt.show()
```

```python
# prompt: can you evaluate the clustering performance using metrics such as silhouette score or DaviesBouldin index?

import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score

features_for_evaluation = df.drop(columns=['cluster'])

# Check if the number of clusters is greater than 1, as Silhouette score is not defined for k=1
if optimal_clusters > 1:
    silhouette_avg = silhouette_score(features_for_evaluation, clusters)
    print(f"\nSilhouette Score: {silhouette_avg:.4f}")
else:
    print("\nSilhouette Score cannot be calculated for 1 cluster.")

# You can also iterate through different numbers of clusters to see how these metrics change
silhouette_scores = []
range_n_clusters_eval = range(2, 11) # Metrics require at least 2 clusters

# We need to refit KMeans for each number of clusters to get new labels
for num_clusters in range_n_clusters_eval:
    kmeans_eval = KMeans(n_clusters=num_clusters, random_state=42, n_init=10)
    clusters_eval = kmeans_eval.fit_predict(features_for_evaluation) # Use the features for evaluation

    # Calculate Silhouette Score
    silhouette_avg_eval = silhouette_score(features_for_evaluation, clusters_eval)
    silhouette_scores.append(silhouette_avg_eval)

# Plot the Silhouette scores
plt.figure(figsize=(10, 6))
plt.plot(range_n_clusters_eval, silhouette_scores, marker='o', linestyle='--')
plt.title('Silhouette Score vs. Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.xticks(range_n_clusters_eval)
plt.grid(True)
plt.show()
```
