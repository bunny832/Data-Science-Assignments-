### Manual Hyperparameter Tuning

In manual tuning, you define your model with specific hyperparameters, train it, and evaluate its performance. You then repeat this process with different hyperparameter values to find the combination that yields the best results. This approach is straightforward but can be time-consuming for a large number of hyperparameters or a wide range of values.

```python
# Example of Manual Hyperparameter Tuning

# Define a function to create and train a model with given hyperparameters
def train_and_evaluate_model(neurons_layer1, neurons_layer2, learning_rate, X_train, y_train_encoded, X_test, y_test_encoded):
    model = keras.Sequential([
        keras.layers.Dense(neurons_layer1, activation='relu', input_shape=(X_train.shape[1],)),
        keras.layers.Dense(neurons_layer2, activation='relu'),
        keras.layers.Dense(26, activation='softmax')
    ])

    optimizer = keras.optimizers.Adam(learning_rate=learning_rate)
    model.compile(optimizer=optimizer,
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

    history = model.fit(X_train, y_train_encoded, epochs=10, batch_size=32, validation_split=0.2, verbose=0)
    loss, accuracy = model.evaluate(X_test, y_test_encoded, verbose=0)

    return accuracy, loss

# Define the hyperparameters and their values to test
param_combinations = [
    {'neurons_layer1': 64, 'neurons_layer2': 32, 'learning_rate': 0.001},
    {'neurons_layer1': 128, 'neurons_layer2': 64, 'learning_rate': 0.001},
    {'neurons_layer1': 256, 'neurons_layer2': 128, 'learning_rate': 0.001},
    {'neurons_layer1': 128, 'neurons_layer2': 64, 'learning_rate': 0.01},
]

best_accuracy = 0
best_params = {}

print("Performing manual hyperparameter tuning...")

for params in param_combinations:
    accuracy, loss = train_and_evaluate_model(
        params['neurons_layer1'],
        params['neurons_layer2'],
        params['learning_rate'],
        X_train, y_train_encoded, X_test, y_test_encoded
    )

    print(f"Params: {params}, Test Accuracy: {accuracy:.4f}, Test Loss: {loss:.4f}")

    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_params = params

print(f"\nBest parameters found: {best_params}")
print(f"Best test accuracy: {best_accuracy:.4f}")
```
