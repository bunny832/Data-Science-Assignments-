```python
# calculating Mean,Median & Mode for numerical values 
print('mean value for Volume :',df['Volume'].mean())
print('median value for Volume:',df['Volume'].median())
print('mode value for Volume:',df['Volume'].mode())
print('standard deviation for Volume:',df['Volume'].std())

print('mean value for Total Sales Value:',df['Total Sales Value'].mean())
print('median value for Total Sales Value:',df['Total Sales Value'].median())
print('mode value for Total Sales Value:',df['Total Sales Value'].mode())
print('standard deviation for Total Sales Value:',df['Total Sales Value'].std())

print('mean value for Avg Price:',df['Avg Price'].mean())
print('median value for Avg Price:',df['Avg Price'].median())
print('mode value for Avg Price:',df['Avg Price'].mode())
print('standard deviation for Avg Price:',df['Avg Price'].std())
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
print(df.info())

numerical_columns = df.select_dtypes(include=[np.number]).columns.tolist()
print(numerical_columns)
print(df.head())
```

```python
# For Standarization 
for col in numerical_columns:
    mean = df[col].mean()
    std = df[col].std()
    df[col + '_standardized'] = (df[col] - mean) / std
    
print(df.head())
```

```python
# Load the dataset
file_path = 'B:\\sales_data_with_discounts.csv'
df = pd.read_csv(file_path)

# Apply one-hot encoding to the categorical columns
df_encoded = pd.get_dummies(df)

# Convert any boolean columns to integers (0 and 1)
df_encoded = df_encoded.astype(int)

# Display the first few rows of the transformed dataset
print(df_encoded.head())

# Save the transformed dataset to a new CSV file
output_file_path = 'B:\\transformed_sales_data_numeric.csv'
df_encoded.to_csv(output_file_path, index=False)

output_file_path
```
